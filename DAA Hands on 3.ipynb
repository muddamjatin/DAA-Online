```python
import time
import numpy as np
import matplotlib.pyplot as plt


def f(n):
    x = 1
    for i in range(1, n):
        for j in range(1, n):
            x += 1
    return x


n_values = list(range(1, 120))
times = []

for n in n_values:
    start_time = time.time()
    f(n)
    end_time = time.time()
    times.append(end_time - start_time)


plt.figure(figsize=(10, 6))
plt.plot(n_values, times, label="Execution Time")
plt.xlabel("n")
plt.ylabel("Time (seconds)")
plt.title("Time vs n")
plt.grid(True)
plt.show()
```

```python

coefficients = np.polyfit(n_values, times, 2)
polynomial = np.poly1d(coefficients)


n_fit = np.linspace(min(n_values), max(n_values), 100)
time_fit = polynomial(n_fit)

plt.plot(n_fit, time_fit, '-', label=f'Fit: {coefficients[0]:.2e}n^2 + {coefficients[1]:.2e}n + {coefficients[2]:.2e}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
coeffs = np.polyfit(n_values, times, 2)
poly_fit = np.poly1d(coeffs)

fit_times = poly_fit(n_values)

plt.figure(figsize=(10, 6))
plt.plot(n_values, times, label="Original Times", color='blue')
plt.plot(n_values, fit_times, label=f"Polynomial Fit: {coeffs[0]:.2e}n^2 + {coeffs[1]:.2e}n + {coeffs[2]:.2e}", color='red')
plt.xlabel("n")
plt.ylabel("Time (seconds)")
plt.title("Time vs n with Polynomial Fit")
plt.legend()
plt.grid(True)
plt.show()

coeffs
```
